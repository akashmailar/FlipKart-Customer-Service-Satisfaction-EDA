## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# 1Ô∏è‚É£ Handling Customer Feedback Data

df["Customer Remarks"] = df["Customer Remarks"].fillna("No Remarks Provided")
```

```python
# 2Ô∏è‚É£ Handling Order & Product Information

df["Order_id"] =  df["Order_id"].fillna("Unknown Order ID")

df["order_date_time"] =  df["order_date_time"].fillna("Not Available")

df["Product_category"] =  df["Product_category"].fillna("Unknown Category")

df["Item_price"] = df["Item_price"].fillna(df["Item_price"].median())  # Filled with median price
```

```python
# 3Ô∏è‚É£ Handling Customer Location Data

df["Customer_City"] = df["Customer_City"].fillna("Unknown City")
```

```python
# 4Ô∏è‚É£ Handling Connected Handling Time Data
# Droping column as too many missing values, otherwise fill with median

df.drop(columns=["connected_handling_time"], inplace=True)  # Since 99.7% missing
```

```python
# 5Ô∏è‚É£ Convert date columns to datetime format
date_cols = ["order_date_time", "Issue_reported_at", "issue_responded", "Survey_response_Date"]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Save the cleaned dataset

df.to_csv("Cleaned_Customer_support_data.csv", index=False)

print("‚úÖ Missing values handled and dataset cleaned successfully!")
```
